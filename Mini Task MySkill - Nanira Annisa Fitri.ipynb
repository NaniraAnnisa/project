# KNN

## Data Preparation

```python
data_x = data.drop('Outcome', axis=1)
data_x
```

```python
data_y = data["Outcome"]
data_y
```

## Latih Model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.33)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```

```python
# kalau mau tahu seberapa cepat melatihnya
import time
start = time.time()
X_train, X_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.33)
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
end = time.time()
print(f"Model training completed in {end-start}")
```

## Validasi/Evaluasi Model

```python
y_pred = model.predict(X_test)
```

```python
accuracy = model.score(X_test, y_test)
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confussion_result = confusion_matrix(y_test, y_pred)
confussion_result
```
